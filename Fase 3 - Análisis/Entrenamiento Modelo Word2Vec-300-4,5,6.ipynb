### sizemodel 300 window 4

```python
sentences = word2vec.Text8Corpus('CORPUS_DEFINITIVO.txt')
sizeModel = 300
model = word2vec.Word2Vec(sentences, size=sizeModel, window=4, min_count=2, workers=cpu_count())
```

```python
model.wv.vocab
```

```python
len(model.wv.vocab)
```

```python
def getFV(document, model1): #promedio del vector caracteristico
    words=document.split()
    s=np.zeros(sizeModel)
    k=1
    for w in words:
        if w in model1.wv.vocab:
            s=s+model1[w]
            k=k+1
    return s/k
```

```python
model.save('w2v300_4.model')
```
