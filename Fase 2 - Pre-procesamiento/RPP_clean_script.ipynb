```python
def clean_parrafo(contenido):
    if not(isinstance(contenido, float)):
        #parrafo_punt = contenido.translate(str.maketrans('', '', string.punctuation)) #Elimina signos de puntuación
        parrafo_punt = contenido.translate(str.maketrans('', '', string.punctuation+'‘’¡¿“”')) #Elimina signos de puntuación
        parrafo_clean = parrafo_punt.lower() #Cambia todo a minuscula
    else:
        parrafo_clean = np.NaN
    return parrafo_clean

def delete_stopwords(cadena, stopwords):
    if not(isinstance(cadena, float)):
        text = ' '.join([word for word in cadena.split() if word not in stopwords])
    else:
        text = np.NaN
    return text


def clean_dataframe(df_clean, stopwords):
    size = df_clean.shape
    for i in range(size[0]):
        df_clean.at[i, 'Contenido'] = delete_stopwords(clean_parrafo(df_clean['Contenido'][i]),stopwords)
        df_clean.at[i, 'Categoria'] = delete_stopwords(clean_parrafo(df_clean['Categoria'][i]),stopwords)
        df_clean.at[i, 'Autor'] = delete_stopwords(clean_parrafo(df_clean['Autor'][i]),stopwords)
        df_clean.at[i, 'Titulo'] = delete_stopwords(clean_parrafo(df_clean['Titulo'][i]),stopwords)
        df_clean.at[i, 'Resumen'] = delete_stopwords(clean_parrafo(df_clean['Resumen'][i]),stopwords)
  
    return df_clean
```

```python
StopWords = stopwords.words("spanish")
df_RPP_clean3 = clean_dataframe(df_RPP_3meses, StopWords)
df_RPP_clean4 = clean_dataframe(df_RPP_4meses, StopWords)
df_RPP_clean5 = clean_dataframe(df_RPP_5meses, StopWords)
```

```python
df_RPP_clean = pd.concat([df_RPP_clean3, df_RPP_clean4, df_RPP_clean5])
```

```python
df_RPP_clean.drop(df_RPP_clean.columns[0],axis=1,inplace=True)
df_RPP_clean.shape
```
